```
import numpy as np
import matplotlib.pyplot as plt
import csv


x=np.arange(1, 141, 1)
ysp=[]
yssp1=[]
yssp2=[]

with open('sp-ld-mi.txt', 'r') as csvfile:
    plots= csv.reader(csvfile, delimiter=',')
    for row in plots:
        ysp.append(float(row[1]))
with open('ssp1-ld-mi.txt', 'r') as csvfile:
    plots= csv.reader(csvfile, delimiter=',')
    for row in plots:
        yssp1.append(float(row[1]))
with open('ssp2-ld-mi.txt', 'r') as csvfile:
    plots= csv.reader(csvfile, delimiter=',')
    for row in plots:
        yssp2.append(float(row[1]))

plt.grid(axis='y', linestyle='--')

plt.plot(x,np.array(ysp),label = "Sandpile Model")
plt.plot(x,np.array(yssp1), label = "Semi-Sandpile Model I")
plt.plot(x,np.array(yssp2), label = "Semi-Sandpile Model II")

plt.title('Comparison of average number of migrations')

plt.xlabel('Load Size')
plt.ylabel('Average Number of Migrations Per Task')
plt.yticks(np.arange(0, 15, 1))
plt.legend()
from google.colab import files
plt.savefig("migrations.eps", format="eps")
files.download("migrations.eps") 
```

```
import numpy as np
import matplotlib.pyplot as plt
import csv


x=np.arange(1, 141, 1)
ysp=[]
yssp1=[]
yssp2=[]
yws=[]

with open('sp-ld-ft.txt', 'r') as csvfile:
    plots= csv.reader(csvfile, delimiter=',')
    for row in plots:
        ysp.append(float(row[1]))
with open('ssp1-ld-ft.txt', 'r') as csvfile:
    plots= csv.reader(csvfile, delimiter=',')
    for row in plots:
        yssp1.append(float(row[1]))
with open('ssp2-ld-ft.txt', 'r') as csvfile:
    plots= csv.reader(csvfile, delimiter=',')
    for row in plots:
        yssp2.append(float(row[1]))
with open('ws-ld-ft.txt', 'r') as csvfile:
    plots= csv.reader(csvfile, delimiter=',')
    for row in plots:
        yws.append(float(row[1]))       

plt.grid(axis='y', linestyle='--')

plt.plot(x,np.array(ysp),label = "Sandpile Model")
plt.plot(x,np.array(yssp1), label = "Semi-Sandpile Model I")
plt.plot(x,np.array(yssp2), label = "Semi-Sandpile Model II")
plt.plot(x,np.array(yws),'k--', label = "Work Stealing Algorithm")

plt.title('Comparison of Flowtime')

plt.xlabel('Load Size')
plt.ylabel('Flowtime (in Seconds)')
plt.yticks(np.arange(0, 31, 2))
plt.legend()
from google.colab import files
plt.savefig("flowtime.eps", format="eps")
files.download("flowtime.eps") 
```
